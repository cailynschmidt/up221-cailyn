```python
#data dowloaded from tiger/line
caTracts=gpd.read_file('data3/tl_2022_06_tract.zip')
caTracts.head()
```

```python
#subset caTracts where COUNTYFP equals '037'
la_tracts=caTracts[caTracts['COUNTYFP']=='037']
la_tracts.head()
```

```python
# view info file of geopandas dataframe
la_tracts.info()
```

```python
#plot geopandas dataframe
la_tracts.plot(figsize=(12,10))
```

```python
# view nonnull values 
la_tracts.info(verbose=True, show_counts=True)
```

```python
#create new data frame with geoid and geometry
la_tracts=la_tracts[['GEOID', 'geometry']]
la_tracts.head()
```

```python
#rename GEOID to Fips to match other dataset column
la_tracts.rename(columns={'GEOID': 'Fips'}, inplace=True)
la_tracts.head()
```

```python
la_tracts.info()
```

```python
#create a new dataframe based on the join
tracts_educ=pd.merge(la_tracts, df2, on='Fips')
tracts_educ.head()
```

```python
tracts_educ.plot(figsize=(12,10),
                  column='<9th grade',
                  legend=True,
                  scheme='NaturalBreaks',
                legend_kwds={'loc':'center left', 'title': '<9th Grade Educational Attainment in LA County', 'bbox_to_anchor':(1,0.5)})
```

```python
# create new columns for percentages 
tracts_educ['PCT_<9thgrade']= tracts_educ['<9th grade']/tracts_educ['Total']*100
tracts_educ['PCT_9th-12th'] = tracts_educ['9th-12th grade, no diploma']/tracts_educ['Total']*100
tracts_educ['PCT_highschool'] = tracts_educ['High school grad']/tracts_educ['Total']*100
tracts_educ['PCT_Bachelors'] = tracts_educ['Bachelors']/tracts_educ['Total']*100
tracts_educ['PCT_Graduate'] = tracts_educ['Graduate or professional']/tracts_educ['Total']*100
```

```python
tracts_educ.plot(figsize=(12,10),
                 column='PCT_<9thgrade',
                 legend=True, 
                 scheme='equal_interval', 
                legend_kwds={'loc':'center left', 'title': '<9th Grade Educational Attainment in LA County', 'bbox_to_anchor':(1,0.5)})
```

```python
ax = tracts_educ.plot(figsize=(12,10),
                 column='PCT_<9thgrade',
                 legend=True, 
                 scheme='equal_interval', 
                legend_kwds={'loc':'center left', 'title': '<9th Grade Educational Attainment in LA County', 'bbox_to_anchor':(1,0.5)})
#save plot as a PNG file
ax.figure.savefig('PCT_<9thgrade_map.png',bbox_inches='tight')
```

```python
tracts_educ.plot(figsize=(12,10),
                 column='PCT_Graduate',
                 legend=True, 
                 scheme='equal_interval', 
                legend_kwds={'loc':'center left', 'title': 'Graduate Educational Attainment in LA County', 'bbox_to_anchor':(1,0.5)})
```

```python
import folium
```

```python
# plot choropleth over the base map
m = folium.Map(location=[34.2,-118.2],
               zoom_start =9,
               tiles ='CartoDB positron',
               attribution='CartoDB')
#plot chorpleth over the base map
folium.Choropleth(
                    geo_data=tracts_educ, # geo data
                    data=tracts_educ,#data
                    key_on='feature.properties.Fips',
                    columns=['Fips', 'PCT_<9thgrade'],
                    fill_color='BuPu',
                    name='Percent Educational Attainment <9th Grade',
                    line_weight=0.1,
                    fill_opacity=0.8,
                    line_opacity=0.2,
                    legend_name='<9th Grade Education Attainment(2022)').add_to(m)

# Add PCT_Gradte choropleth to FeatureGroup
g = folium.Choropleth(
                    geo_data=tracts_educ, # geo data
                    data=tracts_educ,#data
                    key_on='feature.properties.Fips',
                    columns=['Fips', 'PCT_Graduate'],
                    fill_color='YlGn',
                    name='Percent Educational Attainment Graduate',
                    line_weight=0.1,
                    fill_opacity=0.8,
                    line_opacity=0.2,
                    legend_name='<Graduate Education Attainment(2022)').add_to(m)

folium.LayerControl(position='topright',collapsed=True, autoZIndex=True).add_to(m)
m
```
